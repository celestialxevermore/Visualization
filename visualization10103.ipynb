### 5. plt.subplot2grid (More Complex Arrangement)

```python
fig = plt.figure(figsize=(7,7),facecolor='linen')
fig_ = plt.figure(figsize=(7,7),facecolor='linen')
```

```python
ax1 = plt.subplot2grid((2,1),(0,0),fig=fig)
ax2 = plt.subplot2grid((2,1),(1,0),fig=fig)
fig
```

```python
ax1 = plt.subplot2grid((1,2),(0,0),fig=fig_)
ax2 = plt.subplot2grid((1,2),(0,1),fig=fig_)
fig_
```

```python
fig1 = plt.figure(figsize=(7,7),facecolor='linen')
ax1 = plt.subplot2grid((3,3),(0,0),colspan=2,fig=fig1)
ax2 = plt.subplot2grid((3,3),(1,0),colspan=3,fig=fig1)
ax3 = plt.subplot2grid((3,3),(0,2),colspan=1,fig=fig1)
ax4 = plt.subplot2grid((3,3),(2,0),colspan=1,fig=fig1)
ax5 = plt.subplot2grid((3,3),(2,1),colspan=1,fig=fig1)
ax6 = plt.subplot2grid((3,3),(2,2),colspan=1,fig=fig1)
```

```python
fig2 = plt.figure(figsize=(7,7),facecolor='linen')
ax1 = fig2.add_subplot(311)
ax2 = fig2.add_subplot(312)
ax3 = fig2.add_subplot(337)
ax4 = fig2.add_subplot(338)
ax5 = fig2.add_subplot(339)
```

### 예시

```python
fig = plt.figure(figsize=(7,7),facecolor='linen')


ax1 = plt.subplot2grid((3,3),(0,0),colspan=2,fig=fig)
ax2 = plt.subplot2grid((3,3),(1,0),colspan=2,fig=fig)
ax3 = plt.subplot2grid((3,3),(2,0),colspan=2,fig=fig)
ax4 = plt.subplot2grid((1,3),(0,2),rowspan=2,fig=fig)
```

```python
quiz_fig = plt.figure(figsize=(7,7),facecolor='linen')

ax1 = plt.subplot2grid((3,3),(0,0),colspan=3,fig=quiz_fig)
ax2 = plt.subplot2grid((3,3),(1,0),colspan=2,fig=quiz_fig)
ax3 = plt.subplot2grid((3,3),(1,2),rowspan=2,fig=quiz_fig)
ax4 = plt.subplot2grid((3,3),(2,0),colspan=1,fig=quiz_fig)
ax5 = plt.subplot2grid((3,3),(2,1),colspan=1,fig=quiz_fig)
```
